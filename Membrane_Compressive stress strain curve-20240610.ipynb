Area of the 8mm geomegry = $\pi r^2$ (assume the samples were fully covered; can be recalculated)

Stress values were calculated by divide the axial force by the calcaulated area. 

Strain values were provided by the instrument and absolute. 

```python
strain_dry_film_df = pd.read_excel(Dry_data, sheet_name='Axial - 1', usecols=[10], nrows=2000, skiprows=range(0, 2))
stress_dry_film_df = pd.read_excel(Dry_data, sheet_name='Axial - 1', usecols=[12], nrows=2000, skiprows=range(0, 2))
strain_dry_film = strain_dry_film_df.iloc[:, 0].tolist()
stress_dry_film = stress_dry_film_df.iloc[:, 0].tolist()

strain_wet_film_df = pd.read_excel(Wet_data, sheet_name='Axial - 1', usecols=[10], nrows=2000, skiprows=range(0, 2))
stress_wet_film_df = pd.read_excel(Wet_data, sheet_name='Axial - 1', usecols=[12], nrows=2000, skiprows=range(0, 2))
strain_wet_film = strain_wet_film_df.iloc[:, 0].tolist()
stress_wet_film = stress_wet_film_df.iloc[:, 0].tolist()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
c_dry = 'seagreen'
c_wet = 'navy'
plt.plot(strain_dry_film , stress_dry_film, alpha=1, color=c_dry,  linewidth=3, clip_on=False, label= 'Dry state')
plt.plot(strain_wet_film , stress_wet_film, alpha=1, color=c_wet,  linewidth=3, clip_on=False, label= 'Wet state')

plt.xlabel('Strain, $\gamma$ (%)', fontsize=20)
plt.ylabel('Stress $\sigma$ (kPa)', fontsize=20)
plt.tick_params(axis='y', length=5, width=1, color='black', direction='out', labelsize=18)
plt.tick_params(axis='x', length=5, width=1, color='black', direction='out', labelsize=18)
plt.xlim(0, 12)
plt.ylim(0, 100)
plt.legend(frameon=False)
#plt.savefig("Compressive strength of surfactant membrane.png", dpi=300, bbox_inches='tight', transparent=True)
plt.show()
```
